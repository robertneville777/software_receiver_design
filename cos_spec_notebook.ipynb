# Spectrum of a Periodic Signal

Import appropriate libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

Create time vector with 1 second duration sampled at 300 Hz.

```python
t_sim = 1
fs = 300
x = np.arange(0,t_sim,1/fs)
```

Create time signal, in this case, a sum of two cosine waves at different frequencies.

```python
y = np.sin(2*np.pi*100*x) + 0.005*np.cos(2*np.pi*50*x)
```

Take FFT of y.

```python
yf = 20*np.log10(np.fft.fftshift(np.abs(np.fft.fft(y))))
```

Create frequency axis and plot signal in time and frequency.

```python
xf = np.arange(-1/(2*t_sim)*y.size, 1/(2*t_sim)*y.size, 1/(t_sim))

fig = plt.figure()
fig.subplots_adjust(hspace = 0.75)
ax1 = fig.add_subplot(211) # 211 means, 2-by-1 grid, select plot 1
ax1.plot(x,y)
ax1.set_xlabel('Time (seconds)')
ax1.set_ylabel('Amplitude')
ax1.set_title('Signal')
ax1.grid(True)
ax2 = fig.add_subplot(212) # 212 means, 2-by-1 grid, select plot 2
ax2.plot(xf, yf)
ax2.set_xlabel('Frequency (Hz)')
ax2.set_ylabel('Magnitude (dB)')
ax2.set_title('Spectrum of Signal')
ax2.grid(True)
plt.show()
```
